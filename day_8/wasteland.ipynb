## Part 1

```python
lines = []
directions = None
with open('input') as fp:
    directions = fp.readline()
    fp.readline()
    lines = fp.readlines()
```

```python
from dataclasses import dataclass
from typing import Union

@dataclass
class Node:
    left: Union["Node", str]
    right: Union["Node", str]
```

```python
nodes = {}

for line in lines:
    name = line.split("=")[0].strip()
    left, right = line.split("=")[1].strip().replace("(","").replace(")","").replace(",", " ").split()
    nodes[name] = Node(left, right)
```

```python
directions = [*directions.strip()]
```

```python
steps = 0
node = "AAA"
```

```python
dir_idx = 0
while node != "ZZZ":
    if dir_idx == len(directions):
        dir_idx = 0
    direction = directions[dir_idx]
    if direction == "L":
        node = nodes[node].left
    elif direction == "R":
        node = nodes[node].right
    else:
        print("Error", direction, dir_idx)
        break
    dir_idx += 1
    steps += 1
```

## Part 2

```python
from dataclasses import dataclass
from typing import Optional, Union

@dataclass
class Node:
    left: Union["Node", str]
    right: Union["Node", str]
    cycles_left: Optional[int]

nodes = {}

for line in lines:
    name = line.split("=")[0].strip()
    left, right = line.split("=")[1].strip().replace("(","").replace(")","").replace(",", " ").split()
    nodes[name] = Node(left, right, None)
```

```python
starting_nodes = [node for node in list(nodes) if node.endswith("A")]

curr_node = None
for node in starting_nodes:
    curr_node = node
    dir_idx = 0
    steps = 0
    while curr_node.endswith("Z") is False:
        if dir_idx == len(directions):
            dir_idx = 0
        direction = directions[dir_idx]
        if direction == "L":
            curr_node = nodes[curr_node].left
        elif direction == "R":
            curr_node = nodes[curr_node].right
        else:
            print("Error", direction, dir_idx)
            break
        dir_idx += 1
        steps += 1
    nodes[node].cycles_left = steps
```

```python
cycles = [nodes[node].cycles_left for node in starting_nodes]
```

```python
import math
math.lcm(*cycles)
```
