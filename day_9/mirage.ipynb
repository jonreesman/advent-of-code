# Day 9: Mirage Maintenance

## Part 1

```python
lines = []
with open('input') as fp:
    lines = fp.readlines()
```

```python
sequences = []
for line in lines:
    sequences.append([int(num) for num in line.split()])
sequences
```

```python
for idx, sequence in enumerate(sequences):
    sub_sequences = [sequence]
    while True:
        new_sub = []
        for x, y in zip(sub_sequences[-1], sub_sequences[-1][1:]):
            new_sub.append(y - x)
        sub_sequences.append(new_sub)
        if new_sub.count(0) == len(new_sub):
            break
    for index in reversed(range(1, len(sub_sequences))):
        sub_sequences[index-1].append(sub_sequences[index-1][-1] + sub_sequences[index][-1])
    sequences[idx] = sub_sequences[0]
```

```python
sequences
```

```python
total = 0
for sequence in sequences:
    total += sequence[-1]
```

## Part 2

```python
sequences = []
for line in lines:
    sequences.append([int(num) for num in line.split()])
sequences
```

```python
for idx, sequence in enumerate(sequences):
    sub_sequences = [sequence]
    while True:
        new_sub = []
        for x, y in zip(sub_sequences[-1], sub_sequences[-1][1:]):
            new_sub.append(y - x)
        sub_sequences.append(new_sub)
        if new_sub.count(0) == len(new_sub):
            break
    for index in reversed(range(1, len(sub_sequences))):
        sub_1 = sub_sequences[index-1]
        sub_1.reverse()
        sub_2 = sub_sequences[index]
        sub_2.reverse()
        sub_1.append(sub_sequences[index-1][-1] - sub_sequences[index][-1])
        sub_1.reverse()
        sub_sequences[index-1] = sub_1
    sequences[idx] = sub_sequences[0]
```

```python
total = 0
for sequence in sequences:
    total += sequence[0]
```
