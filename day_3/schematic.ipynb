```python
lines = []
with open('input') as fp:
    lines = fp.readlines()
```

```python
symbols = ['!', '@', "*", "#", "$", "%", "^", "&", "/", "-", "+", "="]
```

```python
line = ".........232.633.......................803..........................361................192............539.................973.221...340....."
```

```python
nums = []
```

```python
def in_symbols(char):
    return char in symbols
```

```python
def check_perp(idl, idc):
    valid = False
    if idl > 0:
        if idc > 0:
            valid = in_symbols(lines[idl-1][idc-1]) if not valid else valid
        if idc < len(lines[idl]) - 1:
            valid = in_symbols(lines[idl-1][idc+1]) if not valid else valid
        valid = in_symbols(lines[idl-1][idc]) if not valid else valid
    if idl < len(lines) - 1:
        if idc > 0:
            valid = in_symbols(lines[idl+1][idc-1]) if not valid else valid
        if idc < len(lines[idl]) - 1:
            valid = in_symbols(lines[idl+1][idc+1]) if not valid else valid
        valid = in_symbols(lines[idl+1][idc]) if not valid else valid
    if idc > 0:
        valid = in_symbols(lines[idl][idc-1]) if not valid else valid
    if idc < len(lines[idl]) - 1:
        valid = in_symbols(lines[idl][idc+1]) if not valid else valid
    return valid
            
        
        
```

```python
for idl, line in enumerate(lines):
    idc = 0
    num_window = []
    while idc < len(line):
        if line[idc].isnumeric():
            num_window.append(idc)
        if (not line[idc].isnumeric() or idc == len(line) - 1) and len(num_window) > 0:
            add = False
            for num in num_window:
                if check_perp(idl, num):
                    add = True
            if add:
                n = ""
                for num in num_window:
                    n += line[num]
                nums.append(int(n))
            num_window = []
        idc += 1
        
```

## Part 2

```python
lines = []
with open('input') as fp:
    lines = fp.readlines()
```

```python
symbol = "*"
already_seen = []
```

```python
def check_num(line_num, char_num):
    if not lines[line_num][char_num].isnumeric():
        return
    if line_num < 0 or line_num > len(lines):
        return
    if char_num < 0 or char_num > len(lines[line_num]):
        return
    if (line_num, char_num) in already_seen:
        return
    left_bound = char_num
    right_bound = char_num
    while left_bound > 0:
        if lines[line_num][left_bound - 1].isnumeric():
            left_bound = left_bound - 1
        else:
            break
    while right_bound < len(lines[line_num]):
        if lines[line_num][right_bound + 1].isnumeric():
            right_bound = right_bound + 1
        else:
            break
    if lines[line_num][left_bound:right_bound+1].isnumeric():
        for i in range(left_bound, right_bound + 1):
            already_seen.append((line_num, i))
        return int(lines[line_num][left_bound:right_bound+1])
    return None
```

```python
sum = 0
for idl, line in enumerate(lines):
    for idc, char in enumerate(line):
        if char == "*":
            nums = []
            nums.append(check_num(idl - 1, idc - 1))
            nums.append(check_num(idl - 1, idc + 1))
            nums.append(check_num(idl - 1, idc))
            nums.append(check_num(idl + 1, idc - 1))
            nums.append(check_num(idl + 1, idc + 1))
            nums.append(check_num(idl + 1, idc))
            nums.append(check_num(idl, idc - 1))
            nums.append(check_num(idl, idc + 1))
            nums = [ num for num in nums if num]
            if len(nums) == 2:
                print(idl, idc, nums)
                sum += nums[0] * nums[1]
            if len(nums) != 0 and len(nums) != 2:
                print(idl, idc, nums, "IGNORED")
            
```
